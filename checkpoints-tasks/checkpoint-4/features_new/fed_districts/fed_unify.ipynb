```python
import glob
import pandas as pd
```

```python
pds = []
csvs = glob.glob('*.csv')
for name in csvs:
    df = pd.read_csv(name).drop('Unnamed: 0', axis='columns')
    df['region'] = name.replace("cutted-", "").replace("-fed-district.csv","")
    pds.append(df)
```

```python
pds[0]
```

```python
pd.concat(pds)
```

```python
pd.concat(pds).to_csv('united_objs.csv')
```
