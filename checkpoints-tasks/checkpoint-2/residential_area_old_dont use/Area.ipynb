```python
from geopy.distance import geodesic
import pandas as pd
from os.path import join as joinpath
```

```python
atm_df = pd.read_csv(joinpath("..", "train_test", "train.csv"))
area_df = pd.read_csv("spb_demo.csv")
area_df
```

```python
peter_atm_df = atm_df[atm_df['address_rus'].str.contains('Петербург', na=False)]
peter_atm_df = peter_atm_df.reset_index(drop=True).rename(columns={"Unnamed: 0": "old_index"})
peter_atm_df
```

```python
MIN_DISTANCE_KM = 200 / 1000

area_col = []

for _, atm in peter_atm_df.iterrows():
    atm_coords = atm['lat'], atm['long']
    total_area = 0
    for _, area in area_df.iterrows():
        area_coords = area['Широта'], area['Долгота']
        area_value = area['Площадь (м. кв.)']
        try:
            distance = geodesic(atm_coords, area_coords).km
            if distance <= MIN_DISTANCE_KM:
                total_area += area_value
                print(area['Адрес'])
        except ValueError:
            pass        
    area_col.append(total_area)

peter_atm_df['area'] = area_col
peter_atm_df[peter_atm_df['area'] > 0]
```
