```python
def get_address(addr):
    base_url = 'https://nominatim.openstreetmap.org/search/'
    url = f"{base_url}{addr}"
    data = requests.get(url=url, params={
        'format': 'json',
        'accept-language': 'ru',
        'countrycodes': 'ru',
    }).json()
    print(".", end="")
    return data
```

```python
def transform_addr(addr):
    parts = [j.strip() for j in addr.split(',')]
    good_parts = []
    addr_flag = False
    
    addr_rest = {
        'корпус': '',
        'дом': '',
        'литера': '',
        'строение': '',
    }
    
    abbrs = {
        'корпус': ('к', 'корп'),
        'дом': ('д'),
        'литера': ('л', 'лит'),
        'строение': ('с', 'стр'),
    }
    
    for p in parts:
        p_low = p.lower()
        if p_low.endswith('санкт-петербург'):
            good_parts = ['Петербург']
            addr_flag = True
        elif addr_flag:
            special_part = False
            sub_parts = p_low.replace('.', ' ').split()
            for elem in addr_rest:
                sub_coords = [j for j in [sub_parts.index(i) + 1 for i in (elem, *abbrs[elem]) if i in sub_parts] if j]
                if sub_coords:
                    special_part = True
                    try:
                        addr_rest[elem] = sub_parts[min(sub_coords)]
                    except IndexError:
                        pass
                    break
            if not special_part:
                addr_parts = [i.strip() for i in p.split(' ')]
                if addr_parts and addr_parts[0].endswith('.'):
                    if not addr_parts[0].startswith('г.'):
                        addr_parts.append(addr_parts[0])
                    addr_parts = addr_parts[1:]
                good_parts.append(' '.join(addr_parts))
    
    #return (good_parts, addr_rest)
    if addr_rest['корпус']:
        addr_rest['корпус'] = f"к{addr_rest['корпус']}"
    addr_parsed = f"{addr_rest['дом']}{addr_rest['корпус']}" # {addr_rest['литера']} -- Литера это сложно #СложныйПитер
    good_parts.append(addr_parsed)
    return ' '.join(good_parts).replace('.', '')
```

```python
collect = []
collect_ids = []
```

```python
df_shuffle = df2.sample(n=50)
for _, i in df_shuffle.iterrows():
    area_id = i['id']
    if area_id in collect_ids:
        continue
    collect_ids.append(area_id)
    addr = i['address']
    taddr = transform_addr(addr)
    print(addr)
    print(taddr)    
    variants = get_address(taddr)
    print(variants)
    collect.append([area_id, addr, taddr, variants])
```

```python
clear_collection = [i[:] for i in collect if i[3]]
for c in clear_collection:
    lat, lon = c[-1][0]['lat'], c[-1][0]['lon']
    c.append(lat)
    c.append(lon)
    c.append(df2[df2['id']==c[0]]['area_residential'].item().replace(',','.'))

clear_cols = [
    'id',
    'Адрес',
    'Упрощенный адрес',
    'Результаты запроса',
    'Широта',
    'Долгота',
    'Площадь (м. кв.)',
]
    
df_cc = pd.DataFrame(clear_collection, columns=clear_cols)
df_cc.to_csv('spb_demo.csv')
df_cc
```
